```python
from pytube import YouTube
```

```python
yt = YouTube('https://www.youtube.com/watch?v=k_EEFSvk4Ac')
```

```python
legenda = yt.captions.get_by_language_code('pt')
```

```python
def criar_legenda(link_video):
    yt = YouTube(link_video)
    legenda = yt.captions.get_by_language_code('pt')
    return legenda.xml_captions
```

```python
yt = criar_legenda('https://www.youtube.com/watch?v=f43O-RJFc80')
```
